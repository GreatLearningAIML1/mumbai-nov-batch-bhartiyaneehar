### 5. Calculating distortion - C distance

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial import distance
```

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

<br>Here Kmeans attempts to inimize distortion defined by the sum of the squared distance between each 
observation and its closest centroid</br>

```python

cluster_range = range( 1, 10)   # expect 3 to four clusters from the pair panel visual inspection hence restricting from 2 to 6
distortion=[]
cluster_errors = []
cluster_sil_scores = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 4)#, random_state=2354)
  clusters.fit(df_z)
  labels = clusters.labels_                     # capture the cluster lables
  centroids = clusters.cluster_centers_         # capture the centroids
  cluster_errors.append( clusters.inertia_ )    # capture the intertia
  #cluster_sil_scores.append(metrics.silhouette_score(df_z, labels, metric='euclidean'))
  distortion.append(sum(np.min(distance.cdist(df_z,centroids,'euclidean'),axis=1))/df_z.shape[0])




# combine the cluster_range and cluster_errors into a dataframe by combining them
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors , "distortion": distortion } )
clusters_df[0:15]
```

```python
import matplotlib.pyplot as plt

plt.plot( clusters_df.num_clusters, clusters_df.distortion, marker = "o" )
```

```python
cluster_range = range( 2, 10)   # expect 3 to four clusters from the pair panel visual inspection hence restricting from 2 to 6
distortion=[]
cluster_errors = []
cluster_sil_scores = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 4)#, random_state=2354)
  clusters.fit(df_z)
  labels = clusters.labels_                     # capture the cluster lables
  centroids = clusters.cluster_centers_         # capture the centroids
  cluster_errors.append( clusters.inertia_ )    # capture the intertia
  cluster_sil_scores.append(metrics.silhouette_score(df_z, labels, metric='euclidean'))
  #distortion.append(sum(np.min(distance.cdist(df_z,centroids,'euclidean'),axis=1))/df_z.shape[0])




# combine the cluster_range and cluster_errors into a dataframe by combining them
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors,"Avg Sil Score":cluster_sil_scores } )
clusters_df[0:15]
```

## Hierarchical Clustering 

### 8. Variable creation

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

```python
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])
```

```python
A=pd.DataFrame(a)
B=pd.DataFrame(b)
C=pd.DataFrame(c)

```

### 9. Combine all three arrays a,b,c into a dataframe

```python
A=A.append(B, ignore_index=True)
```

```python
A=A.append(C, ignore_index=True)
```

```python
A
```

### 10. Use scatter matrix to print all the 3 distributions

```python
plt.scatter(x=A[0],y=A[1])
```

### 11. Find out the linkage matrix

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering #ward_tree
from scipy.spatial.distance import pdist  #Pairwise distribution between data points
```

Use ward as linkage metric and distance as Eucledian

```python
Z = linkage(A, 'ward',metric='euclidean')
c, coph_dists = cophenet(Z , pdist(A))

c
```

```python
Z.shape

#Z[1]
#Z[1:500000]
```

### 12. Plot the dendrogram for the consolidated dataframe

```python
plt.figure(figsize=(15, 15))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold=50,  leaf_font_size=10. )
plt.tight_layout()
```

### 13. Recreate the dendrogram for last 12 merged clusters 

```python
plt.figure(figsize=(15, 15))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold=9.5,  leaf_font_size=10,truncate_mode='lastp' )
plt.tight_layout()
```

Hint: Use truncate_mode='lastp' attribute in dendrogram function to arrive at dendrogram 

### 14. From the truncated dendrogram, find out the optimal distance between clusters which u want to use an input for clustering data

<br>As the line Cuts horizantally at 9.5 we compute the Optimal Distance to be 9.5 </br>

### 15. Using this distance measure and fcluster function to cluster the data into 3 different groups

```python
from scipy.cluster import hierarchy  #Pairwise distribution between data points
```

```python
Fcluster =hierarchy.fcluster(Z,t=9.5, criterion='distance')
Fcluster.shape
```
